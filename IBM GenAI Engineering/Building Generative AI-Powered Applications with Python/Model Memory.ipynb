## 2: Manual Deletion via Finder

If you prefer a visual way to drag files to the Trash:
Open Finder.Press Cmd + Shift + G (Go to Folder).

Paste this exact 

- path: ~/.cache/huggingface/hub/

You will see folders starting with models--. 

Each folder represents a model you’ve downloaded.

Right-click the folder (e.g., models--Salesforce--blip-image-captioning-base) and select Move to 

Trash.Important: Empty your Trash to actually reclaim the SSD space.
