# SQL ORDER BY (ASC & DESC)

Real-world table creation, data insertion, and sorting

## 1. Load SQL Extension

```python
%load_ext sql
```

## 2. Connect to Database

```python
%sql postgresql://localhost/sql_for_ds
```

## 3. Create Real-World Table: sales

```python
%%sql
CREATE TABLE sales (
    sale_id INTEGER PRIMARY KEY,
    product_name TEXT,
    category TEXT,
    price INTEGER,
    quantity INTEGER,
    sale_date DATE
);
```

## 4. Insert Real-World Data

```python
%%sql
INSERT INTO sales VALUES
(1, 'Laptop', 'Electronics', 75000, 2, '2024-01-10'),
(2, 'Phone', 'Electronics', 30000, 5, '2024-01-12'),
(3, 'Shoes', 'Fashion', 4000, 10, '2024-01-15'),
(4, 'Watch', 'Accessories', 8000, 3, '2024-01-18'),
(5, 'Headphones', 'Electronics', 2000, 7, '2024-01-20'),
(6, 'T-shirt', 'Fashion', 1200, 15, '2024-01-22');
```

## 5. ORDER BY ASC (Low to High)

```python
%%sql
SELECT * FROM sales ORDER BY price ASC;
```

## 6. ORDER BY DESC (High to Low)

```python
%%sql
SELECT * FROM sales ORDER BY price DESC;
```

## 7. ORDER BY Multiple Columns

```python
%%sql
SELECT * FROM sales ORDER BY category ASC, price DESC;
```

## 8. Convert to Pandas

```python
import pandas as pd
result = %sql SELECT * FROM sales;
df = result.DataFrame()
df.head()
```

## 9. Pandas Sorting

```python
df.sort_values(by='price', ascending=True)
```

```python
df.sort_values(by='price', ascending=False)
```
